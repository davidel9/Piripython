## Exercise 8 — Handle two different exception types

Complete the code so that it:
- catches `ZeroDivisionError` and prints `"Division by zero is not allowed."`
- catches `TypeError` and prints `"Incompatible types for division."`

Test it with:
- `a = 10, b = 0`
- `a = "10", b = 2`

````python
a = 10
b = 0

try:
    result = a / b
    print(result)
except ZeroDivisionError:
    print("Division by zero is not allowed.")
except TypeError:
    print("Incompatible types for division.")
````

## Exercise 9 — Prevent a `KeyError`

You have a dictionary of subject groups.

Tasks:
1. Run the code and identify the exception.
2. Rewrite it using `try` / `except` so that, if the key is missing, it prints:
   `"Subject not found."`

````python


groups = {"s01": "control",
          "s02": "patient",
          "s03": "control"}

subject = "s04"

print(groups[subject])
````

````python

# New version

groups = {"s01": "control",
          "s02": "patient",
          "s03": "control"}

subject = "s04"

try:
    print(groups[subject])
except KeyError:
    print("Subject not found.")
````

## Final reflection

Answer briefly:

1. What is the difference between a `SyntaxError` and an exception?

A SyntaxError happens before the program runs, while an exception occurs during execution.

2. Why is reading a traceback from bottom to top useful?

The bottom shows the actual error and where it occurred, making it easier to detect the cause.

3. Why is catching a specific exception usually better than writing a generic `except`?

Catching specific exceptions avoids hiding other bugs and lets you handle each error.
